```python
import json

products = open("records.json",'r')
r = products.read()
products.close()

items = json.loads(r)
```

```python
items
```

```python
import time
Entry_date_time = time.ctime()
transactions = {}
total = int(input("Enter no of items:"))
for i in range(1,total+1):
      user_prod  = str(input("Enter the product_Id: "))
      if user_prod not in items.keys() or items[user_prod]['qn'] == 0:
          print("The entered product isn't available right now..")
          items[user_prod]['qn'] = items[user_prod]['qn'] + 75
          continue
      user_quant = int(input("Enter the quantity: "))
      transac_num = i
      print("Product: ", items[user_prod]['name'])
      print("Price: ", items[user_prod]['pr'])
      print("Total Amount: ", items[user_prod]['pr'] * user_quant)

      items[user_prod]['qn'] = items[user_prod]['qn'] - user_quant
      transactions[i]  = {'name': items[user_prod]['name'], 'pr': items[user_prod]['pr'], 'qn': user_quant, 'Entry_date_time': Entry_date_time}
```

```python
upd = json.dumps(items)

products = open("records.json",'w')
products.write(upd)
products.close()

upd_sale = json.dumps(transactions)
sale = open("sales.json",'w')
sale.write(upd_sale)
sale.close()
```
