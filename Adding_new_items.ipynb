**Reading json file**

```python
import json
products = open("records.json",'r')
read = products.read()
products.close()
items = json.loads(read)
```

```python
items
```

```python
Total_num = int(input("Enter total number of items:"))
for i in range(Total_num):
        product_id = str(input("Enter product ID:"))
        product_name = str(input("Enter product name:"))
        product_price = int(input("Enter product price:"))
        product_qty = int(input("Enter quantity:"))
        product_brand = str(input("Enter product brand name:"))
        product_mfg = str(input("Enter product manufacturing date:"))
        items[product_id] = {'name': product_name, 'pr': product_price, 'qn': product_qty , 'brand': product_brand , 'Mfg_date': product_mfg}
js = json.dumps(items)
products = open("records.json",'w')
products.write(js)
products.close()
```

***Adding quantity to the inventory products***

```python
prod_add = str(input("Enter product id:"))
prod_qty_add = int(input("Enter Quantity to be added:"))
items[prod_add]['qn'] = items[prod_add]['qn'] + prod_qty_add
print("Added Succesfully")

js = json.dumps(items)
products = open("record.json",'w')
products.write(js)
products.close()
```
